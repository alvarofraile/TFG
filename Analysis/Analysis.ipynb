## Rendimiento

```python
# [Wins, Draws, Losses]
resultsTraining = [403, 6, 91]

keys = ['Win', 'Draw', 'Loss'] 
  
palette = ["#00B000", "#808080", "#F00000"]
  
# plotting data on chart 
plt.pie(resultsTraining, labels=keys, colors=sns.color_palette(palette, 3), autopct='%.0f%%') 
  
# displaying chart 
plt.show() 
```
